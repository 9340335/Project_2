```python
#Используем ссылку на другой ресурс, т.к. запросы на Википедию блокируются бот-фильтром 
import requests
from bs4 import BeautifulSoup

# URL статьи Википедии о городах-миллионниках
url = "https://ru.ruwiki.ru/wiki/%D0%93%D0%BE%D1%80%D0%BE%D0%B4%D0%B0-%D0%BC%D0%B8%D0%BB%D0%BB%D0%B8%D0%BE%D0%BD%D0%B5%D1%80%D1%8B_%D0%A0%D0%BE%D1%81%D1%81%D0%B8%D0%B8"

# Отправляем запрос
headers = {
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36"
}
response = requests.get(url, headers=headers)

if response.status_code == 200:
    soup = BeautifulSoup(response.content, "html.parser")
    
    # Находим таблицу городов-миллионников
    table = soup.find("table", class_="wikitable")
    if table:
        cities_million = []
        # Проходим по строкам таблицы (пропускаем заголовок)
        for row in table.find_all("tr")[1:]:
            cols = row.find_all("td")
            if len(cols) >= 2:
                city = cols[0].get_text(strip=True)
                # Удаляем ссылки и примечания в скобках
                city = city.split("[")[0].strip()
                cities_million.append(city)
        
    else:
        print("Таблица не найдена. Возможно, изменилась структура страницы.")
else:
    print(f"Ошибка загрузки страницы: {response.status_code}")
```

```python
# текст запроса
# преобразуем список городов-миллионников в строку для передачи в sql-запрос
cities_million_str = "('" + "', '".join([str(item) for item in cities_million]) + "')"
query_5_7 = f'''
SELECT
    a.name, 
    COUNT(v.id) cnt
FROM 
    areas a JOIN vacancies v ON a.id = v.area_id JOIN employers e ON v.employer_id = e.id 
WHERE 
    e.name = 'Яндекс' AND a.name IN {cities_million_str}
GROUP BY a.name
UNION ALL
SELECT 
    'total', 
    COUNT(v.id)
FROM areas a JOIN vacancies v ON a.id = v.area_id JOIN employers e ON v.employer_id = e.id 
WHERE 
    e.name = 'Яндекс' AND a.name IN {cities_million_str}
ORDER BY 2    
'''
```

```python
# результат запроса
yandex_vacancies = pd.read_sql_query(query_5_7, connection, params=[cities_million_str])
yandex_vacancies
```

6. Напишите запрос, позволяющий вычислить, какую зарплату для DS в **среднем** указывают для каждого типа требуемого опыта (уникальное значение из поля *experience*). 

При решении задачи примите во внимание следующее:
1. Рассматриваем только вакансии, у которых заполнено хотя бы одно из двух полей с зарплатой.
2. Если заполнены оба поля с зарплатой, то считаем зарплату по каждой вакансии как сумму двух полей, делённую на 2. Если заполнено только одно из полей, то его и считаем зарплатой по вакансии.
3. Если в расчётах участвует null, в результате он тоже даст null (посмотрите, что возвращает запрос select 1 + null). Чтобы избежать этой ситуацию, мы воспользуемся функцией [coalesce](https://postgrespro.ru/docs/postgresql/9.5/functions-conditional#functions-coalesce-nvl-ifnull), которая заменит null на значение, которое мы передадим. Например, посмотрите, что возвращает запрос `select 1 + coalesce(null, 0)`

Выясните, на какую зарплату в среднем может рассчитывать дата-сайентист с опытом работы от 3 до 6 лет. Результат округлите до целого числа.
